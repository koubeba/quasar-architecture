```python
from pyspark.sql import SparkSession, DataFrame
```

```python
DATA_PATH: str = '/media/maya/Samsung_T5/quasar/quasar-split/'
```

```python
spark = SparkSession.builder \
    .master("local") \
    .appName("Concat parquets") \
    .config("spark.sql.execution.arrow.pyspark.enabled", True) \
    .config("spark.sql.parquet.enableVectorizedReader", True) \
    .config("spark.sql.execution.arrow.maxRecordsPerBatch", 10000) \
    .getOrCreate()
```

```python
df = spark.read.parquet(f'{DATA_PATH}/quasar_1.parquet')

for i in range(2, 11):
    print(f'Union with {i}...')
    df = df.union(spark.read.parquet(f'{DATA_PATH}/quasar_{i}.parquet'))
```
