```python
from pyspark.sql import SparkSession, DataFrame
```

```python
DATA_PATH: str = '/media/maja/Samsung_T51/quasar'
```

```python
spark = SparkSession.builder \
    .master("local") \
    .appName("Concat parquets") \
    .config("spark.sql.execution.arrow.pyspark.enabled", True) \
    .config("spark.sql.parquet.enableVectorizedReader", True) \
    .config("spark.sql.execution.arrow.maxRecordsPerBatch", 10000) \
    .getOrCreate()
```

```python
df = spark.read.parquet(f'{DATA_PATH}/spectra-split/spectra-00.parquet')

for i in range(1, 10):
    print(f'Union with {i}...')
    df = df.union(spark.read.parquet(f'{DATA_PATH}/spectra-split/spectra-0{i}.parquet'))
```
