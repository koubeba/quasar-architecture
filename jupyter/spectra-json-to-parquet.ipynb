```python
import pyspark
from   pyspark.sql.types     import StructType, StructField, StringType, DoubleType, IntegerType, ArrayType
```

Import the data from CSV

```python
sc = pyspark.SparkContext(appName="JSON2Parquet")
sqlContext = pyspark.SQLContext(sc)
```

```python
schema = ["spectraSetOID", "mjd", "plate", "fiber", "z", "zerr", "ra", "dec", "type", "subtype", "name", "size", "spectrum", "continuum"]
```

```python
df = sqlContext.read.json(path="/media/maja/Samsung_T51/quasar/spectra-split/09")
```

```python
df.head()
```

Examine the columns and data types

```python
df.dtypes
```

```python
df = df.withColumn("continuum", split(col("continuum"), ",\s*").cast(ArrayType(DoubleType())).alias("continuum"))
df = df.withColumn("spectrum", split(col("spectrum"), ",\s*").cast(ArrayType(DoubleType())).alias("spectrum"))
```

```python
df.head(2)
```

```python
df.write.parquet("/media/maja/Samsung_T51/quasar/spectra-split/spectra-09.parquet")
```
